```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
df2=pd.read_csv("/content/house_prices.csv")
```

```python
df2.head()
```

```python
df2.shape
```

```python
df2.describe()
```

```python
df2.isnull().sum()
```

```python
df2['bathrooms']=df2['bathrooms'].fillna(df2['bathrooms'].mean())
df2.isnull().sum()
```

```python
q1=df2['price'].quantile(0.25)
q3=df2['price'].quantile(0.75)
iqr=q3-q1
upper=q3+3*iqr
df2=df2[df2['price']<upper]
df2.shape
```

```python
x=df2.drop(columns='price')
y=df2['price']
```

```python
print(x)
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(x)
x_std=scaler.transform(x)
```

```python
print(x_std)
```

```python
x=x_std
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
cf2=RandomForestRegressor(n_estimators=200,max_depth=10,random_state=2)
```

```python
cf2.fit(x_train,y_train)
```

```python
x_train_pred=cf2.predict(x_train)
```

```python
x_train_r2=r2_score(y_train,x_train_pred)
```

```python
print(x_train_r2)
```

```python
x_test_pred=cf2.predict(x_test)
x_test_r2=r2_score(y_test,x_test_pred)
print(x_test_r2)
```
